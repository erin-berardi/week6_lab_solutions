```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Display the 20 latest earthquakes info (date, time, magnitude, and region name) by the EMSC as a pandas dataframe

```python
url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
soup.select('td:nth-child(11)')[0].get_text()
```

```python
soup.select('td:nth-child(12)')[0].get_text()
```

```python
#\31 242996 > td.tabev6 > b > a
```

```python
soup.select('td.tabev6 > b > a')[0].get_text()
```

```python
date = []
magnitude = []
location = []

for i in range(21):
    date.append(soup.select('td.tabev6 > b > a')[i].get_text())
    magnitude.append(soup.select('td:nth-child(11)')[i].get_text())
    location.append(soup.select('td:nth-child(12)')[i].get_text())
```

```python
date
```

```python
magnitude
```

```python
location
```

```python
quakes = pd.DataFrame({"Date and Time":date,
                       "Magnitude":magnitude,
                       "Location":location})
```

```python
quakes
```

```python
quakes['Date and Time'] = pd.to_datetime(quakes['Date and Time'])
```
